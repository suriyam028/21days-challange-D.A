```python
print(feedback_data)
avg = sum(feedback_data['Rating']) / len(feedback_data['Rating'])
print("Average Rating:", avg)
longest = ""
max_words = 0

for feedback in feedback_data['Feedback']:
    word_count = len(feedback.split())
    if word_count > max_words:
        max_words = word_count
        longest = feedback

print("Longest Feedback:", longest)
unique_words = set()

for feedback in feedback_data['Feedback']:
    words = feedback.split()
    unique_words.update(words)

print("Unique words:", unique_words)
```

```python
def count_word_in_feedbacks(word):
    count = 0
    for feedback in feedback_data['Feedback']:
        if word.lower() in feedback:
            count += 1
    return count

print("Good count:", count_word_in_feedbacks("good"))
print("Poor count:", count_word_in_feedbacks("poor"))
print("Excellent count:", count_word_in_feedbacks("excellent"))
```

```python
def clean_text(text):
    text = text.replace('.', '').replace(',', '').replace('!', '').replace('?', '')
    text = ' '.join(text.split())   # removes extra spaces
    text = text.strip()             # remove leading/trailing spaces
    text = text.lower()
    return text

# Apply cleaning
for i in range(len(feedback_data['Feedback'])):
    feedback_data['Feedback'][i] = clean_text(feedback_data['Feedback'][i])
```

```python
n = int(input("How many feedbacks do you want to add? "))

for i in range(n):
    name = input("Enter name: ")
    feedback = input("Enter feedback: ")
    rating = int(input("Enter rating (1-5): "))

    feedback_data['S_No'].append(len(feedback_data['S_No']) + 1)
    feedback_data['Name'].append(name)
    feedback_data['Feedback'].append(feedback)
    feedback_data['Rating'].append(rating)
```
