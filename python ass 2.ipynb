```python
score = int(input("Enter your score (0 to 10): "))

if score > 7:
    print("Above Average: Excellent performance! Keep it up.")

elif score >= 4 and score <= 7:
    print("Average: Good effort! Keep practicing.")

elif score < 4:
    print("Below Average: Need to improve your performance.")

else:
    print("Invalid score! Please enter a value between 0 and 10.")
```

```python
set1 = {1, 3, 5, 7, 9}
set2 = {2, 3, 5, 8, 10}

print("Set1:", set1)
print("Set2:", set2)
print("Union:", set1.union(set2))
print("Intersection:", set1.intersection(set2))
```

```python
student_marks = {
    "Arun": 75,
    "Bala": 82,
    "Charan": 90,
    "Divya": 68,
    "Elena": 85
}

print("Student Marks:", student_marks)
print("Marks of Bala:", student_marks["Bala"])
student_marks["Janani"] = 80
print("After adding Janani:", student_marks)
student_marks["Arun"] = 82
print("After updating Arun:", student_marks)
print("Student Names:", student_marks.keys())
print("Marks:", student_marks.values())
print("Key-Value Pairs:", student_marks.items())
```

```python
name_list.append("Yazhini")
print("After append:", name_list)
age_list.insert(2, 30)
print("After insert:", age_list)
name_list.remove("Yazhini")
print("After remove:", name_list)
age_list.pop()
print("After pop:", age_list)
age_list.extend([29, 30, 26])
print("After extend:", age_list)
age_list.sort(reverse=True)
print("Sorted descending:", age_list)
print("Maximum age:", max(age_list))
print("Minimum age:", min(age_list))
print("Sum of ages:", sum(age_list))
```
